# 2. CleaningData

```python
# Rename columns
pugetsound_tides.rename(columns={'/data/item/pred_in_cm': 'Height (cm)', '/data/item/pred_in_ft': 'Height (ft)', '/stationid': 'Tide_Station_ID', '/stationname': 'Tide_Station_Name'}, inplace=True)
```

```python
pugetsound_tides.head()
```

```python
# Reorder columns
desired_order = ['Dive_Site', 'Date', 'Day_of_Week', 'Time', 'Event', 'Height (cm)', 'Height (ft)', 'Tide_Station_Name', 'Tide_Station_ID', 'Latitude', 'Longitude']
pugetsound_tides = pugetsound_tides[desired_order]
```

```python
pugetsound_tides.shape
```

```python
# Convert the standardized dates to the desired format ("YYYY-MM-DD") for pugetsound_currents
pugetsound_tides['Date'] = pd.to_datetime(pugetsound_tides['Date'], errors='coerce')
```

```python
# Convert the "Time" column to datetime time format
pugetsound_tides['Time'] = pd.to_datetime(pugetsound_tides['Time'], format='%H:%M').dt.time
```

```python
# Convert the "Time" column to datetime time format
pugetsound_tides['Time'] = pd.to_datetime(pugetsound_tides['Time'], format='%H:%M:%S').dt.time
```

```python
# Convert the "Time" column to datetime format and remove seconds
pugetsound_tides['Time'] = pd.to_datetime(pugetsound_tides['Time'], format='%H:%M:%S').dt.strftime('%H:%M')
```

```python
# Dictionary to map abbreviated day names to full day names
day_name_mapping = {
    'Mon': 'Monday',
    'Tue': 'Tuesday',
    'Wed': 'Wednesday',
    'Thu': 'Thursday',
    'Fri': 'Friday',
    'Sat': 'Saturday',
    'Sun': 'Sunday'
}

# Apply the mapping to the "Day_of_Week" column
pugetsound_tides['Day_of_Week'] = pugetsound_tides['Day_of_Week'].map(day_name_mapping)
```

```python
# Dictionary to map event abbreviations to full names
event_mapping = {
    'L': 'Low',
    'H': 'High'
}

# Apply the mapping to the "Event" column
pugetsound_tides['Event'] = pugetsound_tides['Event'].map(event_mapping)
```

```python
# Convert Tide_Station_ID to object
pugetsound_tides['Tide_Station_ID'] = pugetsound_tides['Tide_Station_ID'].astype(str)
```

```python
# Remove leading spaces from column names
pugetsound_tides.columns = pugetsound_tides.columns.str.strip()
```
