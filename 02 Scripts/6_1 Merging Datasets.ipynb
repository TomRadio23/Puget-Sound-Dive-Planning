# Merge Datasets

```python
# Define the common columns to merge on
merge_columns = ['Dive_Site', 'Date', 'Day_of_Week', 'Time', 'Event', 'Latitude', 'Longitude']
```

```python
# Merge pugetsound_currents_clean and pugetsound_tides_clean using a full outer join
merged_data = pugetsound_currents_clean.merge(pugetsound_tides_clean, on=merge_columns, how='outer')
```

```python
# Convert the "Time" column to datetime time format
merged_data['Time'] = pd.to_datetime(merged_data['Time'], format='%H:%M').dt.time
```

```python
merged_data.info()
```

```python
# Convert the "Time" column to string format
merged_data['Time'] = merged_data['Time'].astype(str)

# Convert the "Time" column to datetime format and extract the time component
merged_data['Time'] = pd.to_datetime(merged_data['Time'], format='%H:%M:%S').dt.time

# Display the updated data types
print(merged_data.dtypes)
```

```python
# Sort the merged_data dataframe
sorted_data = merged_data.sort_values(by=['Dive_Site', 'Date', 'Time'])
```

```python
# Rename the sorted_data dataset to pugetsound_tides_currents
pugetsound_tides_currents = sorted_data.copy()
```

```python
pugetsound_tides_currents.head(50)
```

```python
# Reorder the columns in the desired order
column_order = ['Dive_Site', 'Date', 'Day_of_Week', 'Time', 'Event', 'Current_Speed_(knots)', 'Height (cm)', 'Height (ft)', 'Current_Station_Name', 'Current_Station_ID', 'Tide_Station_Name', 'Tide_Station_ID', 'Latitude', 'Longitude']
pugetsound_tides_currents = pugetsound_tides_currents[column_order]
```

```python
# Reset the index to get consecutive index numbers
pugetsound_tides_currents.reset_index(drop=True, inplace=True)
```
